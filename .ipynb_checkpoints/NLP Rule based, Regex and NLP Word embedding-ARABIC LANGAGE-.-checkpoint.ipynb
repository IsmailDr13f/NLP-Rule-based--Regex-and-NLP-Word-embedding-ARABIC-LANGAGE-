### <b>Part 1:Rule Based NLP and Regex:</b>

<b>Target:</b> generate bill from a text given by user.

```python
string_input="I bought three Samsung smartphones 150 $ each, four kilos of fresh banana for 1,2 dollar a kilogram and one Hamburger with 4,5 dollar"
```

```python
import re
# Pattern to extract the quantity:
pattern_Q = r'\b(?:one|two|three|four|five|six|seven|eight|nine|ten|eleven|twelve|thirteen|fourteen|fifteen|sixteen|seventeen|eighteen|nineteen|twenty|thirty|forty|fifty|sixty|seventy|eighty|ninety|hundred|thousand)(?:\s+(?:and\s+)?(?:one|two|three|four|five|six|seven|eight|nine))?'
Quantity = re.findall(pattern_Q,string_input)
Quantity
```

```python
# Transfering numbers from words to numerics:
# !pip install word2number
from word2number import w2n
Quant = [w2n.word_to_num(wrd) for wrd in Quantity]
Quant
```

```python
# Pattern to extract the price:
pattern_P = r"[-+]?(?:\d*\,*\d+)"
Price_U = re.findall(pattern_P,string_input)
Price_U = [float(i.replace(",", ".")) for i in Price_U]
Price_U
```

```python
# Pattern to extract prodects:
pattern_Pro = re.compile(r'(?:one|two|three|four|five|six|seven|eight|nine|ten)\s+(?:kilo of|kilos of fresh|kilos of|(\w+))\s+(\w+)\b', re.IGNORECASE)

# Find all matches in the text
Prodects = pattern_Pro.findall(string_input)
Prodects
```

```python
# Filtring extracted words by type (Prodect always a Propre noun or noun)
import spacy
nlp = spacy.load("en_core_web_sm")
def word_type_filtre(tpl):
    phrase = nlp(' '.join(tpl))
    nouns=[wrd.text for wrd in phrase if wrd.pos_ == 'PROPN' or wrd.pos_ == 'NOUN']
    return nouns        
```

```python
Prods=[' '.join(word_type_filtre(elem)) for elem in Prodects]
Prods
```

```python
Quantity
```

```python
import pandas as pd

data={'Prodect': Prods,'Quantity': Quant,'Unit_Price': Price_U}
df= pd.DataFrame(data)
df['Total Price']=df['Quantity']*df['Unit_Price']
df
```

### <b>Target_1:</b> Apply one hot encoding, bag of words and TF-IDF technics.

#### #One-hot encoding:

```python
# Create vocabulary set
vocab_set = set()
for sent in sentences_:
    kalimat = sent.split()
    for kalima in kalimat:
        vocab_set.add(kalima)
print(vocab_set)

kalimat_ = []
for sent in sentences_:
    kalimat = sent.split()
    for kalima in kalimat:
        kalimat_.append(kalima)
print(kalimat_)
```

```python
# dictionary to map words to ints
kalima_ila_raqm = {kalima: raqm for raqm, kalima in enumerate(vocab_set)}
kalima_ila_raqm
```

```python
import numpy as np
# Create binary vectors for each word in each sentence
vectors_ = []
for sent in sentences_:
    kalimat = sent.split()
    sent_vect = []
    for kalima in kalimat:
        bin_vect = np.zeros(len(vocab_set))
        bin_vect[kalima_ila_raqm[kalima]] = 1
        sent_vect.append(bin_vect)
    vectors_.append(sent_vect)
vectors_
```

```python
# Results:
for i in range(len(sentences_)):
    print(f"العبارة {i + 1}:")
    for j in range(len(vectors_[i])):
        print(f"{sentences_[i].split()[j]}: {vectors_[i][j]}")
    
```

### <b>Target_2:</b> Apply one Word2Vec Approach.

#### #CBOW:

```python
import regex as re
import numpy as np
import matplotlib.pyplot as plt
```

```python
sentences_
```

```python
print(kalimat_)
```

```python
# Create vocabulary 
print(vocab_set)
```

```python
vocab_size = len(vocab_set)
embed_dim = 10
context_size = 2

#Dictionnary
print(kalima_ila_raqm)

raqm_ila_kalima = {value:key for key, value in kalima_ila_raqm.items()}
print(raqm_ila_kalima)
```

```python
data__ = []
for i in range(2, len(kalimat_) - 2):
    context = [kalimat_[i - 2], kalimat_[i - 1], kalimat_[i + 1], kalimat_[i + 2]]
    target = kalimat_[i]
    data__.append((context, target))
print(data__[:5])
```

```python
# embedings
embeddings = np.random.random_sample((vocab_size, embed_dim))
```

```python
# linear model
def linear(m, theta):
    w = theta
    return m.dot(w)
    
# log softmax
def log_softmax(x):
    e_x = np.exp(x - np.max(x))
    return np.log(e_x / e_x.sum())
    
def NLLLoss(logs, targets):
    out = logs[range(len(targets)), targets]
    return -out.sum()/len(out)
    
def log_softmax_crossentropy_with_logits(logits,target):
    out = np.zeros_like(logits)
    out[np.arange(len(logits)),target] = 1
    softmax = np.exp(logits) / np.exp(logits).sum(axis=-1,keepdims=True)
    return (- out + softmax) / logits.shape[0]
    
def forward(context_idxs, theta):
    m = embeddings[context_idxs].reshape(1, -1)
    n = linear(m, theta)
    o = log_softmax(n)
    return m, n, o
    
def backward(preds, theta, target_idxs):
    m, n, o = preds
    dlog = log_softmax_crossentropy_with_logits(n, target_idxs)
    dw = m.T.dot(dlog)
    return dw
    
def optimize(theta, grad, lr=0.03):
    theta -= grad * lr
    return theta
    
# init thetas
theta = np.random.uniform(-1, 1, (2 * context_size * embed_dim, vocab_size))
epoch_losses = {}

for epoch in range(80):
    losses = []
    for context, target in data__:
        context_idxs = np.array([kalima_ila_raqm[w] for w in context])
        preds = forward(context_idxs, theta)
        target_idxs = np.array([kalima_ila_raqm[target]])
        loss = NLLLoss(preds[-1], target_idxs)
        losses.append(loss)
        grad = backward(preds, theta, target_idxs)
        theta = optimize(theta, grad, lr=0.03)
    epoch_losses[epoch] = losses
    
ix = np.arange(0,80)
fig = plt.figure()
fig.suptitle('Epoch/Losses', fontsize=20)
plt.plot(ix,[epoch_losses[i][0] for i in ix])
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Losses', fontsize=12)
```

```python
def predict(words):
    context_idxs = np.array([kalima_ila_raqm[w] for w in words])
    preds = forward(context_idxs, theta)
    word = raqm_ila_kalima[np.argmax(preds[-1])]
    return word , np.argmax(preds[-1])
    
predict(['ونسألك', 'يا', 'دوام', 'العطايا'])
```

#### #Skip Gram:

```python
import regex as re
import numpy as np
import matplotlib.pyplot as plt

# Tokens
print(kalimat_)
print(kalima_ila_raqm)
print(raqm_ila_kalima)
```

```python
# Generate training data
def generate_training_data(tokens, word_to_id, window_size):
    X, Y = [], []
    for i in range(len(tokens)):
        nbr_inds = list(range(max(0, i - window_size), i)) + list(range(i + 1, min(len(tokens), i + window_size + 1)))
        for j in nbr_inds:
            X.append(word_to_id[tokens[i]])
            Y.append(word_to_id[tokens[j]])
    return np.array(X), np.array(Y)
```

```python
x, y = generate_training_data(kalimat_, kalima_ila_raqm, 3)
print(x)
print(y)
```

```python
def expand_dims(x, y):
    x = np.expand_dims(x, axis=0)
    y = np.expand_dims(y, axis=0)
    return x, y
```

```python
x, y = expand_dims(x, y)
print(x)
print(y)
```

```python
# forward propagation
def init_parameters(vocab_size, emb_size):
    wrd_emb = np.random.randn(vocab_size, emb_size) * 0.01
    w = np.random.randn(vocab_size, emb_size) * 0.01
    return wrd_emb, w

def softmax(z):
    return np.divide(np.exp(z), np.sum(np.exp(z), axis=0, keepdims=True) + 0.001)

def forward(inds, params):
    wrd_emb, w = params
    word_vec = wrd_emb[inds.flatten(), :].T
    z = np.dot(w, word_vec)
    out = softmax(z)
    cache = inds, word_vec, w, z
    return out, cache

# cost funtion
def cross_entropy(y, y_hat):
    m = y.shape[1]
    cost = -(1 / m) * np.sum(np.sum(y_hat * np.log(y + 0.001), axis=0, keepdims=True), axis=1)
    return cost

# dervation of softmax
def dsoftmax(y, out):
    dl_dz = out - y
    return dl_dz

def backward(y, out, cache):
    inds, word_vec, w, z = cache
    wrd_emb, w = params
    dl_dz = dsoftmax(y, out)
    # deviding by the word_vec length to find the average
    dl_dw = (1/word_vec.shape[1]) * np.dot(dl_dz, word_vec.T)
    dl_dword_vec = np.dot(w.T, dl_dz)
    return dl_dz, dl_dw, dl_dword_vec

def update(params, cache, grads, lr=0.03):
    inds, word_vec, w, z = cache
    wrd_emb, w = params
    dl_dz, dl_dw, dl_dword_vec = grads
    wrd_emb[inds.flatten(), :] -= dl_dword_vec.T * lr
    w -= dl_dw * lr
    return wrd_emb, w
```

```python
# training
vocab_size = len(raqm_ila_kalima)
m = y.shape[1]
y_one_hot = np.zeros((vocab_size, m))
y_one_hot[y.flatten(), np.arange(m)] = 1
y = y_one_hot
batch_size=256
embed_size = 50
params = init_parameters(vocab_size, 50)
costs = []

for epoch in range(5000):
    epoch_cost = 0
    batch_inds = list(range(0, x.shape[1], batch_size))
    np.random.shuffle(batch_inds)
    for i in batch_inds:
        x_batch = x[:, i:i+batch_size]
        y_batch = y[:, i:i+batch_size]
        pred, cache = forward(x_batch, params)
        grads = backward(y_batch, pred, cache)
        params = update(params, cache, grads, 0.03)
        cost = cross_entropy(pred, y_batch)
        epoch_cost += np.squeeze(cost)

    costs.append(epoch_cost)
    if(epoch % 250 == 0):
        print("Cost after epoch {}: {}".format(epoch, epoch_cost))

ix = np.arange(0,5000)
fig = plt.figure()
fig.suptitle('Epoch/Losses', fontsize=20)
plt.plot(ix,[costs[i] for i in ix])
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Losses', fontsize=12)

# test
x_test = np.arange(vocab_size)
x_test = np.expand_dims(x_test, axis=0)
softmax_test, _ = forward(x_test, params)
top_sorted_inds = np.argsort(softmax_test, axis=0)[-4:,:]

for input_ind in range(vocab_size):
    input_word = raqm_ila_kalima[input_ind]
    output_words = [raqm_ila_kalima[output_ind] for output_ind in top_sorted_inds[::-1,input_ind]]
    print("{}'s skip-grams: {}".format(input_word, output_words))
```

#### #Glove:

```python
import torch
import torchtext.vocab as vocab
# Load the pre-trained GloVe model trained on Arabic text
glove = vocab.GloVe(name='twitter.27B', dim=200)  # You might need to adjust the dimension depending on the available model
```

```python
# TEST:
# Define Arabic word pairs to compute similarity for
word_pairs = [('الصدقات', 'العطايا'), ('الناجحين', 'الفائزين'), ('العطايا', 'والنعم')]

# Compute similarity for each pair of Arabic words
for pair in word_pairs:
    vec1, vec2 = glove[pair[0]], glove[pair[1]]
    similarity = torch.dot(vec1, vec2) / (torch.norm(vec1) * torch.norm(vec2))
    print(f"Similarity between '{pair[0]}' and '{pair[1]}' using GloVe: {similarity:.3f}")

```

### <b>Target_4:</b> Plot all the encoded / vectorized vectors by using Tsne Algorithm:

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Convert embeddings lists to NumPy arrays
embeddings1_np = np.array(embeddings1)
embeddings2_np = np.array(embeddings2)

# Combine the embeddings arrays
embeddings_np = np.concatenate((embeddings1_np, embeddings2_np), axis=0)
labels = arabic_words1 + arabic_words2

# Apply t-SNE to reduce the dimensionality to 2D
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(embeddings_np)
```

```python
# Plot the embeddings
plt.figure(figsize=(10, 8))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])
for i, word in enumerate(labels):
    plt.annotate(word, xy=(embeddings_2d[i, 0], embeddings_2d[i, 1]), fontsize=8)
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.title("t-SNE Plot of Arabic Word Embeddings (Fasttext)")
plt.grid(True)
plt.show()
```
